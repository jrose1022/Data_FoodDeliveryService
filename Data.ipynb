# **DATA**

```python
import pandas as pd
import numpy as np

np.random.seed(42)

rows = 1000

data = {
    "order_id": np.arange(1, rows+1),
    "customer_id": np.random.randint(1000, 2000, rows),
    "restaurant_name": np.random.choice(["Jollibee", "McDo", "KFC", "Mang Inasal", "Chowking"], rows),
    "food_category": np.random.choice(["Fast Food", "Filipino", "Chinese", "Dessert"], rows),
    "order_date": pd.date_range(start="2026-04-01", periods=rows, freq="H"),
    "delivery_time_minutes": np.random.randint(20, 90, rows),
    "order_status": np.random.choice(["Delivered", "Late", "Cancelled"], rows),
    "payment_method": np.random.choice(["Cash", "GCash", "Card"], rows),
    "order_amount": np.round(np.random.uniform(100, 1000, rows), 2),
    "delivery_fee": np.round(np.random.uniform(20, 100, rows), 2),
    "rider_id": np.random.randint(1, 100, rows),
    "city": np.random.choice(["Manila", "Cavite", "Laguna", "Pasig"], rows)
}

df = pd.DataFrame(data)

df.head()
```

# **DATAFRAME FILTERING**

*1. Orders from April 1 to April 4, 2026*

```python
df[(df['order_date'] >= '2026-04-01') & (df['order_date'] <= '2026-04-04')]
```

*2. Late orders only*

```python
df[df['order_status'] == 'Late']
```

*3. Orders above ₱500*

```python
df[df['order_amount'] > 500]
```

*4. Orders paid using GCash*

```python
df[df['payment_method'] == 'GCash']
```

*5. Orders in Cavite*

```python
df[df['city'] == 'Cavite']
```

6. Fast food category *only*

```python
df[df['food_category'] == 'Fast Food']
```

*7. Delivery time greater than 60 minutes*

```python
df[df['delivery_time_minutes'] > 60]
```

*8. Delivered orders only*

```python
df[df['order_status'] == 'Delivered']
```

# **AGGREGATION & GROUPING**

*1. Orders per restaurant*

```python
df.groupby('restaurant_name')['order_id'].count()
```

*2. Average delivery time per city*

```python
df.groupby('city')['delivery_time_minutes'].mean()
```

*3. Total revenue per restaurant*

```python
df.groupby('restaurant_name')['order_amount'].sum()
```

*4. Average order amount per payment method*

```python
df.groupby('payment_method')['order_amount'].mean()
```

*5. Orders per status*

```python
df.groupby('order_status')['order_id'].count()
```

*6. Average delivery fee per city*

```python
df.groupby('city')['delivery_fee'].mean()
```

*7. Orders per food category*

```python
df.groupby('food_category')['order_id'].count()
```

*8. Average delivery time per restaurant*

```python
df.groupby('restaurant_name')['delivery_time_minutes'].mean()
```
