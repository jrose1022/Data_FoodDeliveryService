```python
import pandas as pd
import numpy as np

np.random.seed(42)

rows = 1000

data = {
    "order_id": np.arange(1, rows+1),
    "customer_id": np.random.randint(1000, 2000, rows),
    "restaurant_name": np.random.choice(["Jollibee", "McDo", "KFC", "Mang Inasal", "Chowking"], rows),
    "food_category": np.random.choice(["Fast Food", "Filipino", "Chinese", "Dessert"], rows),
    "order_date": pd.date_range(start="2026-04-01", periods=rows, freq="H"),
    "delivery_time_minutes": np.random.randint(20, 90, rows),
    "order_status": np.random.choice(["Delivered", "Late", "Cancelled"], rows),
    "payment_method": np.random.choice(["Cash", "GCash", "Card"], rows),
    "order_amount": np.round(np.random.uniform(100, 1000, rows), 2),
    "delivery_fee": np.round(np.random.uniform(20, 100, rows), 2),
    "rider_id": np.random.randint(1, 100, rows),
    "city": np.random.choice(["Manila", "Cavite", "Laguna", "Pasig"], rows)
}

df = pd.DataFrame(data)

df.to_csv("food_delivery_data.csv", index=False)

df.head()
```
